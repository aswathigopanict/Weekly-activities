## **Bitwise operators.**

### AND operation



```python
a=10
b=4
a & b
```

a = 10 = 1010 (Binary)
b = 4 =  0100 (Binary

a & b = 1010
         &
        0100
      = 0000
      = 0 (Decimal)

###  OR operation


```python
a=10
b=4
(a | b )
```

a = 10 = 1010 (Binary)
b = 4 =  0100 (Binary

a | b = 1010
         |
        0100
      = 1110
      = 14 (Decimal)

### XOR operation


```python
a=10
b=4
a ^ b 
```

a = 10 = 1010 (Binary)
b = 4 =  0100 (Binary

a & b = 1010
         ^
        0100
      = 1110
      = 14 (Decimal)

#### How it works
a = 10 = 1010 (Binary)

~a = ~1010
   = -(1010 + 1)
